```python
import pandas as pd

rat_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=rat_cols, usecols=range(3), encoding="ISO-8859-1")

mov_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=mov_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)

ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=50)
corrMatrix.head()
```

Now listing my manually inserted ratings

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

Now I will list the possible recommendations of movies based on the movies that are similar to the ones I rated.

There will be a list of movies for every movie I rated, all coming from the original correlation matrix. I will scale those scores based on how good score I gave those movies, therefore the higher rated movie will have higher weight.

```python
simMovies = pd.Series()
for i in range(0, len(myRatings.index)):
    print ("Finding similar movies for " + myRatings.index[i] + "...")   
    sims = corrMatrix[myRatings.index[i]].dropna()   # List of similar movies to the one in question
    sims = sims.map(lambda x: x * myRatings[i])   # Scaling the movies correlation with my rating
    simMovies = simMovies.append(sims)   # Add the score to the list of movies that have correlation with the one I rated
    
#Results:
print ("\nResults:")
simMovies.sort_values(inplace = True, ascending = False)
print (simMovies.head(10))
```

```python
simMovies = simMovies.groupby(simMovies.index).sum()
simMovies.sort_values(inplace = True, ascending = False)
simMovies.head(10)
```

```python
filteredSims = simMovies.drop(myRatings.index)
filteredSims.head(10)
```
